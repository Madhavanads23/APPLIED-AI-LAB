```python
import numpy as np

def step(x):
    return 1 if x >= 0 else 0

# User inputs
n = int(input("Enter number of inputs: "))
lr = float(input("Enter learning rate: "))
epochs = int(input("Enter number of epochs: "))

weights = np.array(list(map(float, input("Enter initial weights: ").split())))
bias = float(input("Enter bias: "))

# Input data
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

def train_gate(gate_name, y, w, b):
    w = w.copy()

    for _ in range(epochs):
        for xi, target in zip(X, y):
            output = step(np.dot(xi, w) - b)
            error = target - output
            w += lr * error * xi
            b -= lr * error   # ðŸ”´ IMPORTANT CHANGE

        print(f"\n{gate_name} Gate:")0.1 0.3
        
        print("Weights:", w)
        print("Bias:", b)

        for x in X:
            print(x, "->", step(np.dot(x, w) - b))

# AND gate
y_and = np.array([0, 0, 0, 1])
train_gate("AND", y_and, weights, bias)

# OR gate
y_or = np.array([0, 1, 1, 1])
train_gate("OR", y_or, weights, bias)
```
